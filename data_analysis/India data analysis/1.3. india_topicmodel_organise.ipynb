#### Organising data in preparation for close reading grounded coding


```python
from bertopic import BERTopic

topic_inmodel=BERTopic.load("topic_inmodel_merge")
```

```python
# fetching inbody from excel file previously saved

import pandas as pd

indf=pd.read_excel('in_input_data.xlsx')

input_data = pd.read_excel('in_input_data.xlsx')

intitle=input_data['title']
inbody=input_data['body']
indate=input_data['date']
insource=input_data['source']
```

#### Dataframe for topics and their representative documents

```python
topics_info_withoutliers=topic_inmodel.get_topic_info()

# Dropping the outliers from the dataframe 
topics_info = topics_info_withoutliers.drop(0)  # Drop the first row
topics_info = topics_info.reset_index(drop=True)  # Reset the index numbering

topics_info
```

```python
# Getting a dataframe with three representative documents 

inrepdocs = []
for n in range(0, len(topics_info)):
    new_df = {}
    new_df['topic'] = n
    new_df['Doc1'] = topic_inmodel.get_representative_docs(n)[0]
    new_df['Doc2'] = topic_inmodel.get_representative_docs(n)[1]
    new_df['Doc3'] = topic_inmodel.get_representative_docs(n)[2]
    inrepdocs.append(new_df)

topics_repdocs = pd.DataFrame(inrepdocs)
topics_repdocs
```

```python
# Combining the two dataframes

topics_newdf = pd.concat([topics_info,topics_repdocs],axis=1) 
topics_alldf = topics_newdf.drop(['Representative_Docs', 'topic'], axis=1)

# If you want to modify the DataFrame in-place, use the inplace parameter
# topics_newdf.drop(['Representation', 'Representative_Docs', 'topic'], axis=1, inplace=True)

# Display the modified DataFrame
topics_alldf
```

```python
# save as excel file

topics_alldf.to_excel('in_newsbody_repdocs.xlsx', index=False)
```

#### Dataframe for all articles and their related information

```python
indocs_all=topic_inmodel.get_document_info(inbody)
```

```python
# Get document info and combine with other data, drop unnecessary columns, rename, and reorder
indocs_allinfo = (pd.concat([indocs_all, indf], axis=1)
                  .drop(['Top_n_words', 'Document'], axis=1)
                  .rename(columns={'title': 'Headline',
                                   'body': 'Body',
                                   'date': 'Date',
                                   'source': 'Source'})
                  [['Headline', 'Body', 'Topic', 'Name', 'Representation', 'Date', 'Source', 
                    'Representative_Docs', 'Probability', 'Representative_document']])

# Display the final DataFrame
indocs_allinfo
```

```python
indocs_allinfo.to_excel('in_newsbody_allinfo.xlsx', index=False)
```
